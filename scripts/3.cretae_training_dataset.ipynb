```python
import polars as pl

eng_corpus = pl.read_parquet("../data/outputs_similarity_matches/corpus_eng.parquet")
jap_corpus = pl.read_parquet("../data/outputs_similarity_matches/corpus_jap.parquet")
```

```python
# x data 

# english

eng_x = (
    eng_corpus.select(
        (pl.col("title") + " " + pl.col("abstract")).alias("text")
    )
    .to_series()
    .to_list()
)

# japanese 

jap_x = (
    jap_corpus.select(
        (pl.col("title_ja") + " " + pl.col("abstract_ja")).alias("text")
    )
    .to_series()
    .to_list()
)

print(len(eng_x)), print(eng_x[0:5])
print(len(jap_x)), print(jap_x[0:5])
```

```python
# y data

eng_y = (
    eng_corpus.select(
    pl.col('label').alias("label")
    )
    .to_series()
    .to_list()
)

jap_y = (
    jap_corpus.select(
    pl.col('label').alias("label")
    )
    .to_series()
    .to_list()
)

print(len(eng_y)), print(eng_y)
print(len(jap_y)), print(jap_y)
```

```python
# save to disk

import mpu 

mpu.io.write("../data/for_analysis/eng_x.pickle", eng_x)
mpu.io.write("../data/for_analysis/jap_x.pickle", jap_x)

mpu.io.write("../data/for_analysis/eng_y.pickle", eng_y)
mpu.io.write("../data/for_analysis/jap_y.pickle", jap_y)
```
